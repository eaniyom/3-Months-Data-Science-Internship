### 1. Number Theory
Number theory is a branch of mathematics that deals with the properties and relationships of numbers, particularly integers. It is a fundamental field of study that explores the nature of numbers and their patterns. Number theory has applications in various areas, including cryptography, computer science, and physics.

###### Here are some key concepts and topics in number theory:

1. Prime Numbers: Prime numbers are positive integers greater than 1 that have no divisors other than 1 and themselves. They play a crucial role in number theory. The prime factorization of a number expresses it as a product of prime numbers.

2. Divisibility and Factors: Divisibility refers to the property of one number being divisible by another without leaving a remainder. Factors are the numbers that divide a given number evenly. The study of divisibility and factors helps understand the structure of numbers.

3. Modular Arithmetic: Modular arithmetic is a system of arithmetic where numbers "wrap around" after reaching a certain value called the modulus. It has applications in cryptography, computer science algorithms, and number theory itself.

4. Congruence Relations: Congruence is a relation between two integers where their difference is divisible by a given modulus. It generalizes the notion of equality. For example, 15 ≡ 3 (mod 12) means that 15 and 3 leave the same remainder when divided by 12.

5. Diophantine Equations: Diophantine equations are polynomial equations that seek integer solutions. The most famous example is Fermat's Last Theorem, which states that there are no whole number solutions to the equation x^n + y^n = z^n for n > 2.

6. The Fundamental Theorem of Arithmetic: This theorem states that every positive integer greater than 1 can be uniquely factorized into a product of prime numbers. It establishes the fundamental role of prime numbers in the study of integers.

7. Number Systems: Number theory also explores various number systems, such as rational numbers (fractions), real numbers, and complex numbers. Each system has its unique properties and relationships with the integers.

8. Continued Fractions: Continued fractions are a way of representing real numbers as an infinite sequence of nested fractions. They provide an alternative representation of numbers and have connections to various areas of mathematics.



Number theory goes all the way back to ancient times, when different number systems were studied. 

The number systems in this study is largely divided into two classes

* Real Number Systems 
* Complex Number Systems 


### A. Real Numbers
Real numbers include rational as well as irrational numbers. In
practicality, when you are doing any data science work you can treat

The whole numbers are the numbers without fractions and it is a collection of positive integers and zero. Adding to natural numbers, the concept of “0” was later accepted; we call these “whole numbers.”

## 2. Order of Operations

In data science, as in mathematics, it is important to follow the order of operations to ensure the correct evaluation of expressions. The order of operations determines the sequence in which different operations are performed. The standard order of operations used in data science is as follows: You can remember the order of operations by the mnemonic device PEMDAS (Please Excuse My Dear Aunt Sally): 

* Parentheses: Operations enclosed in parentheses have the highest precedence. Evaluate expressions inside parentheses first.

* Exponents: Evaluate any exponentiation (raising to a power) operations next. For example, calculate squares or higher powers of numbers.

* Multiplication and Division: Perform multiplication and division operations from left to right. These operations have equal precedence. If there are multiple multiplication or division operations, evaluate them in the order they appear in the expression.

* Addition and Subtraction: Perform addition and subtraction operations from left to right. Similar to multiplication and division, these operations have equal precedence. If there are multiple addition or subtraction operations, evaluate them in the order they appear in the expression.

It's important to note that when two operations have the same precedence (such as addition and subtraction, or multiplication and division), the left-to-right order is followed. This is known as the associativity of operators.

Additionally, some data science libraries or programming languages may have additional functions or operators with specific rules for evaluation. It's crucial to refer to the documentation of the specific library or programming language you are using to understand any additional rules or functions available.

## 4. Functions
Functions are expressions that define relationships between two or more
variables. More specifically, a function takes input variables (also called
domain variables or independent variables), plugs them into an expression,
and then results in an output variable (also called dependent variable).


Given the function;
<center>

For any given x-value, we solve the expression with that x to find y. When x = 1, then y = 3. When x = 2, y = 5. When x = 3, y = 7 and so on, as shown in the Table below;

|x|2x+1|y|
|--|--|--|
|0|2(0)+1|1|

## 8. Euler’s Number and Natural Logarithms

Euler's number, denoted as "e," is a mathematical constant that arises in various areas of mathematics and science. It is an irrational number, meaning it cannot be expressed as a fraction of two integers, and its decimal representation is infinite and non-repeating.

Euler's number has a value approximately equal to 2.71828, although its decimal representation continues indefinitely. The value of e is commonly used as the base of the natural logarithm, denoted as ln(x). The natural logarithm is the inverse function of exponentiation with base e.

Euler's number is named after the Swiss mathematician Leonhard Euler, who made significant contributions to various branches of mathematics. The number e was first introduced by Jacob Bernoulli in the early 18th century while studying compound interest and continuous growth.

The significance of Euler's number arises from its many applications in mathematics, physics, and engineering. Some key applications include:

- Compound Interest: Euler's number is used in the context of continuous compounding, where the value of an investment or debt continuously grows or decays over time. The formula A = P*e^(rt) calculates the final amount (A) based on the principal (P), interest rate (r), and time (t).

- Calculus and Differential Equations: Euler's number appears in calculus and differential equations, particularly in the study of exponential and logarithmic functions. It plays a crucial role in solving differential equations and modeling various natural phenomena.

- Probability and Statistics: Euler's number is involved in the definition of the exponential distribution, which is used to model events with a constant rate of occurrence over time. It also arises in the context of continuous probability distributions, such as the normal distribution.

- Complex Analysis: Euler's number is linked to the complex exponential function, which has a profound connection with trigonometry and complex numbers. Euler's formula, e^(ix) = cos(x) + i*sin(x), unifies exponential functions, trigonometric functions, and imaginary numbers.

- Physics: Euler's number appears in various physical laws and phenomena, including radioactive decay, electrical circuits, growth models, and wave equations. It provides a mathematical foundation for describing exponential growth and decay processes.

a = P(1+(r/n)^(n*t))

```python
# Calculating compound interest in Python
from math import exp
p = 100
r = .20
t = 2.0
n = 24
a = p * (1 + (r/n))**(n * t)
print(a) # prints 148.69146179463576
```

## 9. Limits
In Mathematics, a limit is defined as a value that a function approaches the output for the given input values. 

The way we read this is “as x approaches infinity, the function 1/x
approaches 0 (but never reaches 0).” You will see this kind of “approach but never touch” behavior a lot, especially when we dive into derivatives and integrals.

```python
# Using SymPy to calculate limits
from sympy import *

x = symbols('x')
f = 1 / x
result = limit(f, x, oo)
print(result) # 0
```

As you have seen, we discovered Euler’s number e this way too. It is the
result of forever extending n into infinity for this function: 

```python
from sympy import *

n = symbols('n')
f = (1 + (1/n))**n
result = limit(f, n, oo)
print(result) # E
print(result.evalf()) # 2.71828182845905
```

## 10. Derivatives
A derivative tells the slope of a function, and it is useful to measure the rate of change at any point in a function.

Let's find the derrivative of the function 

```python
# Calculating a derivative in SymPy
from sympy import *
# Declare 'x' to SymPy
x = symbols('x')

# Now just use Python syntax to declare function
f = x**2

# Calculate the derivative of the function
dx_f = diff(f)
print(dx_f) # prints 2*x
```

Find the derivative at x = 2.0

```python
# A derivative calculator in Python
def f(x):
    return x**2

def dx_f(x):
    return 2*x

slope_at_2 = dx_f(2.0)
print(slope_at_2) # prints 4.0
```

```python
# 6. For the function f (x) = 3x2 + 1 what is the slope at x = 3?

from sympy import *

# Declare 'x' to SymPy
x = symbols('x')

# Now just use Python syntax to declare function
f = 3*x**2 + 1

# Calculate the derivative of the function
dx_f = diff(f)

print('Derivative of {0} is given as {1}'.format(f, dx_f)) # prints 6*x
print()
print('The slope is:', dx_f.subs(x,3)) # 18
```

```python
# 7. For the function f (x) = 3x2 + 1 what is the area under the curve for x between 0 and 2?

from sympy import *

# Declare 'x' to SymPy
x = symbols('x')

# Now just use Python syntax to declare function
f = 3*x**2 + 1

# Calculate the integral of the function with respect to x for the area between x = 0 and 2
area = integrate(f, (x, 0, 2))
print(area) # prints 10
```
