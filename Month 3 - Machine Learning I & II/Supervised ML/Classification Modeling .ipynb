### 1. Introduction to Machine Learning
A machine learning model is a program that can find patterns or make decisions from a previously unseen dataset. For example, in natural language processing, machine learning models can parse and correctly recognize the intent behind previously unheard sentences or combinations of words. In image recognition, a machine learning model can be taught to recognize objects - such as cars or dogs. A machine learning model can perform such tasks by having it 'trained' with a large dataset. During training, the machine learning algorithm is optimized to find certain patterns or outputs from the dataset, depending on the task. The output of this process - often a computer program with specific rules and data structures - is called a machine learning model.


#### What is a Machine Learning Algorithm
A machine learning algorithm is a mathematical method to find patterns in a set of data. Machine Learning algorithms are often drawn from statistics, calculus, and linear algebra

#### What is Machine model training
The process of running a machine learning algorithm on a dataset (called training data) and optimizing the algorithm to find certain patterns or outputs is called model training. The resulting function with rules and data structures is called the trained machine learning model.

### 3. Classification models
Classification models are a type of machine learning model that is used to predict the class of an input data point. The class of an input data point is a category or group that the data point belongs to. For example, a classification model could be used to predict whether an email is spam or not spam, or whether a patient has cancer or not cancer.


### 4. Types of Classification Algorithms
There are many different types of classification models, each with its own strengths and weaknesses. Some of the most popular classification models include:

1. Logistic regression: This is a simple but powerful model that can be used for both binary and multi-class classification.
2. Decision trees: Decision trees are a versatile model that can be used for a wide variety of classification tasks.
3. Random forests: Random forests are an ensemble model that combines multiple decision trees to improve accuracy.
4. Support vector machines: Support vector machines are a powerful model that can be used for both binary and multi-class classification.
5. K-nearest neighbors: K-nearest neighbors is a simple but effective model that can be used for both classification and regression tasks.


The best classification model for a particular task will depend on the nature of the data and the desired accuracy.

Here are some of the steps involved in building a machine learning model:
* Data collection: The first step is to collect a dataset of labeled data. This dataset should include examples of each of the classes that the model will be trained to predict.
* Data preprocessing: The next step is to preprocess the data. This may involve cleaning the data, removing outliers, and converting the data to a format that the model can understand.
* Model selection: Once the data has been preprocessed, the next step is to select a classification model. There are many different models to choose from, so it is important to select a model that is appropriate for the task at hand.
* Model training: Once a model has been selected, the next step is to train the model. This involves feeding the model the labeled data and allowing the model to learn the relationships between the features and the classes.
* Model evaluation: Once the model has been trained, the next step is to evaluate the model. This involves feeding the model new data that it has not seen before and seeing how well the model predicts the classes of the new data.
* Model deployment: Once the model has been evaluated and found to be satisfactory, the model can be deployed. This means that the model can be used to predict the classes of new data.

```python
#1. import necessary libraries
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pandas.plotting import scatter_matrix
```

```python
#2. Creating a function for loading data
def load_dataset():
    url = "https://storage.googleapis.com/qwasar-public/track-ds/iris.csv"
    dataset = pd.read_csv(url)
    return dataset
load_dataset()
```

```python
#3. Function for statistical summary

def summarize_dataset(dataset):
    # The dimension of the dataset
    print("The dimension:")
    print(dataset.shape)
    print("")
    
    # The first 10 rows of the dataset
    print("The first 10 rows of the dataset: ")
    print(dataset.head(10))
    print("")
    
    # The statistical summary
    print("The statistical summary:")
    print(dataset.describe())
    print("")
    
    # Class group
    print("Classification:")
    print(dataset.groupby('class').size())
    
summarize_dataset(load_dataset())
```

```python
#4. Visualization for univariate plot
def print_plot_univariate(dataset):
    # Making histogram plot
    dataset.hist()
    plt.show()
    
# Visualization for bivariate plot    
def print_plot_multivariate(dataset):
    # Scatter_matrix
    scatter_matrix(dataset)
    plt.show()
        
print_plot_univariate(load_dataset())
print_plot_multivariate(load_dataset())
```

```python
#5.Model Training  
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.tree import DecisionTreeClassifier

def my_print_and_test_models(dataset):
    array = dataset.values
    X = array[:,0:4]
    y = array[:,4]
    
    # Data split into train and validation data
    X_train, X_validation, Y_train, Y_validation = train_test_split(X, y, test_size=0.20, random_state=42)
    
    # Define models
    models = [
        ('DecisionTree', DecisionTreeClassifier(random_state=42))
    ]

    # Train and evaluate each model
    for model_name, model in models:
        cv_results = cross_val_score(model, X_train, Y_train, cv=5)
        print(f'{model_name}: {cv_results.mean():.6f} ({cv_results.std():.6f})')

# Usage example
my_print_and_test_models(load_dataset())
```
